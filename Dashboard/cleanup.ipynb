```python
#Import packages
from path import Path
import pandas as pd
import matplotlib.pyplot as mp
import seaborn as sb
```

```python
data = Path("Datasets1/WorldHappinessReport2021_headers.csv")
df = pd.read_csv(data)
df.head()
```

```python
df.shape
```

```python
df.info()
```

```python
df.describe()
```

```python
clean_df = df.drop(columns=['upperwhisker', 'lowerwhisker', 'std_error_ladder_score', 'logged_GDP_per_capita', 'healthy_life_expectancy', 'life_choices', 'generosity', 'corruption', 'dystopia_ladder_score', 'social_support'])
clean_df.head()
```

```python
clean_col_df = clean_df.rename(columns={'country_name':'Country', 'regional_indicator':'Region', 'ladder_score':'Happiness Score', 
                        'GDP_per_capita_Explained':'GDP per Capita X', 'social_support_Explained':'Social Support X', 'healthy_life_expectancy_Explained':'Healthy Life Expectancy X',
                        'life_choices_Explained':'Life Choices X', 'generosity_Explained':'Generosity X', 'corruption_Explained':'Corruption X', 'Dystopia_residual':'Dystopia + Residual'})
clean_col_df.head()
```

```python
clean_col_df.corr()
```

```python
corrheatmap = sb.heatmap(clean_col_df.corr(), cmap="YlGnBu", annot=True)
mp.show()
```

```python
clean_col_df.sort_values('Happiness Score', ascending=False)
clean_col_df.head()
```

```python
clean_bar_df = clean_col_df.drop(columns=["Happiness Score"])
clean_bar_df.head()
```

```python
clean_bar_df.plot(x='Country', kind='barh', width=0.5, figsize=(15,45), stacked=True, title='Happiness Score by Country')
# stacked_bar_df.drop(columns="Happiness Score")
```

```python
#Find the mean of the scores for each region of the world. 
region_df = clean_col_df.groupby('Region').mean()
region_df.head(10)
```
