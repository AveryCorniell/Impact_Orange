```python
#Import EDA
import numpy as np # linear algebra
import pandas as pd # data processing, CSV file I/O (e.g. pd.read_csv)

# import sklearn library
import sklearn.datasets as datasets
import sklearn.preprocessing as preprocessing
import sklearn.model_selection as model_selection
import sklearn.metrics as metrics
import sklearn.linear_model as linear_model
from sklearn.metrics import mean_squared_error
from sklearn.metrics import max_error
import sklearn
from sklearn import linear_model

# import statistic library
from scipy import stats
import statsmodels.api as sm

#Import Graph
import seaborn as sns
import matplotlib.pyplot as plt
import plotly.express as px
```

```python
#Import Data
df = pd.read_csv("data/world-happiness-report.csv")
df_2021 = pd.read_csv("data/world-happiness-report-2021.csv")
```

```python
# Get a list of all countries
df_2021['Country name'].unique()
```

```python
# count regional indicator
sns.countplot(df_2021["Regional indicator"])
plt.xticks(rotation=60) # show axis name nicely
plt.show()
```

```python
df_2021_happiest_unhappiest = df_2021[(df_2021.loc[:, "Ladder score"] > 7.5) | (df_2021.loc[:, "Ladder score"] < 4)]
sns.barplot(x = "Ladder score", y="Country name", data=df_2021_happiest_unhappiest, palette="coolwarm")
plt.title("Happiest and Unhappiest Countries in 2021")
plt.show()
```

```python
# Happiest and unhappiest countries (multi year)

df_happiest_unhappiest = df[(df.loc[:, "Life Ladder"] > 7.5) | (df.loc[:, "Life Ladder"] < 4)]
sns.barplot(x = "Life Ladder", y="Country name", data=df_happiest_unhappiest, palette="coolwarm")
plt.title("Happiest and Unhappiest Countries in 2008 - 2020")
plt.show()
```
