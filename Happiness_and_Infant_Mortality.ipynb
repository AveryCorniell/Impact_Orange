```python
# Dependencies and Setup
import pandas as pd
import matplotlib.pyplot as plt
import numpy as np

# Files to load
world_happiness = "Datasets1/WorldHappinessReport_headers.csv"
infant_mortality_rate = "Datasets1/InfantMortalityRate_headers.csv"

# Read the world happiness and infant mortality rate files into a dataframe
world_happiness_df = pd.read_csv(world_happiness)
infant_mortality_df = pd.read_csv(infant_mortality_rate)
```

```python
# Rename the country_name column in world_happiness_df to "country"
world_happiness_df.rename(columns = {'country_name' : 'country'}, inplace = True)
```

```python
# Combine the data into a single dataset
happiness_df = pd.merge(world_happiness_df, infant_mortality_df, how="left", on=["country", "country"])

# Display the data table for preview
happiness_df.head()
```

```python
# Drop null values
happiness_df.dropna()
```
